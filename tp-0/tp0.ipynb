# Image processing with Python

## Reading, Converting from RGB to Greyscale and Showing images
### Using PILLOW

```python
import cv2
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Reading image
img = Image.open('../resources/lena.png')
img
```

```python
# Converting to grayscale
img_gray = img.convert('L')
img_gray
```

```python
# Showing two images 
plt.figure()
f, axarr = plt.subplots(1,2)

axarr[0].set_title('Original Image')
axarr[0].imshow(np.array(img))

axarr[1].set_title('Grayscale Image')
axarr[1].imshow(np.array(img_gray), cmap='gray')
```

```python
# Saving the image
img_gray.save('./lena_gray_pillow.png')
```

### Using OpenCV

```python
# Reading the image 
img = cv2.imread('../resources/lena.png')
```

```python
# Showing the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# Image Shape
img.shape
```

```python
# Converting image to array
img_array = np.array(img)
```

```python
# Reshape image
img_reshape = img.reshape(img.shape[2], img.shape[0], img.shape[1])
print(img_reshape.shape)
```

```python
# Conversion to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Showing both images 
plt.figure()
f, axarr = plt.subplots(1, 2)

axarr[0].set_title('Original Image')
axarr[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

axarr[1].set_title('Grayscale Image')
axarr[1].imshow(img_gray, cmap='gray')
plt.show()
```

```python
# Saving the image
cv2.imwrite('./lena_gray_cv2.png', img_gray)

# Saving the figure from plot 
f.savefig("figure.png")
```
